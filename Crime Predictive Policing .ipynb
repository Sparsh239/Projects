```python
from scipy.stats import chi2_contingency
from scipy.stats import chi2
display(Markdown("## Location Description Includes "))
location_description_value_counts = pd.DataFrame(df['Location Description'].value_counts().reset_index())
location_description_value_counts.columns = ['Location Description', 'Count']
location_description_value_counts['Percentage'] = (location_description_value_counts['Count']/location_description_value_counts['Count'].sum()) * 100
display(HTML(location_description_value_counts.head(10).to_html()))
fig = px.bar(location_description_value_counts, x="Location Description", y="Count", color="Location Description", title="Location Description Bar Chart")
fig.show()
fig = px.pie(location_description_value_counts, values='Percentage', names='Location Description', title="Location Description Pie Chart",color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()

display(Markdown("Chi Square Test"))
data_crosstab = pd.crosstab(df['Warning'], df['Location Description'], margins = False) 
stat, p, dof, expected = chi2_contingency(data_crosstab)
print("\nDegree of Freedom:", dof)
     # interpret test-statistic
prob = 0.95
critical = chi2.ppf(prob, dof)
print('probability=%.3f, critical=%.3f, stat=%.3f' % (prob, critical, stat))
     # interpret p-value
alpha = 1.0 - prob
print('significance=%.3f, p=%.3f' % (alpha, p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (fail to reject H0)')
```

```python
arrest_value_counts = pd.DataFrame(df['Arrest'].value_counts().reset_index())
arrest_value_counts.columns = ['Arrest', 'Count']
arrest_value_counts['Percentage'] = (arrest_value_counts['Count']/arrest_value_counts['Count'].sum()) * 100
display(HTML(arrest_value_counts.head(10).to_html()))
fig = px.bar(arrest_value_counts, x="Arrest", y="Count",color_discrete_sequence=px.colors.sequential.RdBu, title="Arrest Bar Chart")
fig.show()
fig = px.pie(arrest_value_counts, values='Percentage', names='Arrest', title="Arrest Pie Chart",color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()

display(Markdown("#### Chi Square Test : Arrest and Crime Type "))
data_crosstab = pd.crosstab( df['Arrest'],df['Warning'], margins = False) 
display(data_crosstab)
stat, p, dof, expected = chi2_contingency(data_crosstab)
print("\nDegree of Freedom:", dof)
     # interpret test-statistic
prob = 0.95
critical = chi2.ppf(prob, dof)
print('probability=%.3f, critical=%.3f, stat=%.3f' % (prob, critical, stat))
     # interpret p-value
alpha = 1.0 - prob
print('significance=%.3f, p=%.3f' % (alpha, p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (fail to reject H0)')

display(Markdown("#### Chi Square Test : Arrest and Location Description "))
data_crosstab = pd.crosstab( df['Arrest'],df['Location Description'],margins = False) 
display(data_crosstab)
stat, p, dof, expected = chi2_contingency(data_crosstab)
print("\nDegree of Freedom:", dof)
     # interpret test-statistic
prob = 0.95
critical = chi2.ppf(prob, dof)
print('probability=%.3f, critical=%.3f, stat=%.3f' % (prob, critical, stat))
     # interpret p-value
alpha = 1.0 - prob
print('significance=%.3f, p=%.3f' % (alpha, p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (fail to reject H0)')
```

### Model Deployment 

```python
from sklearn.model_selection  import train_test_split
# split the data with 50% in each set
X = np.array(features)

y = np.array(analysis_dataset[['Categorical_Response__labels']])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.4, random_state= 80)
```

```python
print(X_train.shape, X_test.shape)
```

```python
print(y_train.shape, y_test.shape)
```

```python
y_train = y_train.ravel()
y_test = y_test.ravel()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
SVC_model = SVC()
KNN_model = KNeighborsClassifier(n_neighbors=5)
DTC_model = DecisionTreeClassifier()
GNB_model = GaussianNB()
LDA_model = LinearDiscriminantAnalysis()
```

```python
display(Markdown("# Support Vetor Machine"))
SVC_model.fit(X_train, y_train)
SVC_prediction = SVC_model.predict(X_test)
print(accuracy_score(SVC_prediction, y_test))
print(confusion_matrix(SVC_prediction, y_test))
print(classification_report(SVC_prediction, y_test))
```

```python
display(Markdown("# K Nearest Neighbours"))
KNN_model = KNeighborsClassifier(n_neighbors=5)
KNN_model.fit(X_train, y_train)
KNN_prediction = KNN_model.predict(X_test)
print(accuracy_score(KNN_prediction, y_test))
print(confusion_matrix(KNN_prediction, y_test))
print(classification_report(KNN_prediction, y_test))
```

```python
display(Markdown("# Decision Tree Classifier"))
DTC_model = DecisionTreeClassifier()
DTC_model.fit(X_train, y_train)
DTC_prediction = DTC_model.predict(X_test)
print(accuracy_score(DTC_prediction, y_test))
print(confusion_matrix(DTC_prediction, y_test))
print(classification_report(DTC_prediction, y_test))
```
